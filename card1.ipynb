```python
# convert 'reg_preference_for_trad' to numerical
def score_to_numeric(x):
    if x=='一线城市':
        return 0
    if x=='二线城市':
        return 1
    if x=='三线城市':
        return 2
    if x=='其他城市':
        return 3
    if x=='境外':
        return 4
    
```

```python
df1['location'] = df1['reg_preference_for_trad'].apply(score_to_numeric)
df1
```

```python
df2= df1.drop('reg_preference_for_trad', axis=1) 
```
