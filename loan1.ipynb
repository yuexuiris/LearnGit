```python
# convert 'reg_preference_for_trad' to numerical
def char_to_numeric(x):
    if x=='一线城市':
        return 0
    if x=='二线城市':
        return 1
    if x=='三线城市':
        return 2
    if x=='其他城市':
        return 3
    if x=='境外':
        return 4
    
```

```python
df1['location'] = df1['reg_preference_for_trad'].apply(char_to_numeric)
df1
```

```python
# the distribution of numerical variables
def get_feature_groups():
    """ Returns a list of numerical and categorical features,
    excluding SalePrice and Id. """
    # Numerical Features
    num_features = df1.select_dtypes(include=['int64','float64']).columns
    num_features = num_features.drop(['custid']) # drop custid 

    # Categorical Features
    cat_features = df1.select_dtypes(include=['object']).columns
    cat_features =cat_features.drop(['loans_latest_time']) # drop loans_latest_time
    return list(num_features), list(cat_features)

num_features, cat_features = get_feature_groups()
```

```python
# Distribution plots of all numerical features
f = pd.melt(df1, value_vars=sorted(num_features))
g = sns.FacetGrid(f, col='variable', col_wrap=4, sharex=False, sharey=False)
g = g.map(sns.distplot, 'value')
```
